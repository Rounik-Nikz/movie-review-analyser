```python
#Importing Required Libraries
import pandas as pd
import numpy as np
import csv
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
```

```python
#Reading the data which is provided in the same directory with name 'movie_sentiments_data.csv' and storing it in df variable
df = pd.read_csv("movie_sentiments_data.csv",quoting=csv.QUOTE_NONE,on_bad_lines='skip')


#Printing the shape of the data
print(df.shape)

#Printing top 5 datapoints
df.head()
```

```python
#creating a new column "Category" which represent 1 if the sentiment is positive or 0 if it is negative
df['Category'] = df['sentiment'].apply(lambda x: 1 if x =='positive' else 0)
```

```python
#checking the distribution of 'Category' and observing whether the Target labels are balanced or not.
df['Category'].value_counts()
```

```python
#Performing the 'train-test' splitting with test size of 20%

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.review.fillna(' '), df.Category, test_size=0.2)
```

```python
#Creating a pipeline object
clf = Pipeline([
    ('vectorizer', CountVectorizer()),                                                    #initializing the vectorizer
    ('random_forest', (RandomForestClassifier(n_estimators=50, criterion='entropy')))      #using the RandomForest classifier
])



#fitting it with X_train and y_train
clf.fit(X_train, y_train)


#Getting the predictions for X_test and storing it in y_pred
y_pred = clf.predict(X_test)


#Printing the classfication report
print(classification_report(y_test, y_pred))
```
